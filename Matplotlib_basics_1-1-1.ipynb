```python
#importing the pyplot module as plt 
import matplotlib.pyplot as plt

square_values = [1, 4 , 9 , 16 , 25 , 36 , 49 , 64 , 81 ]

#plot the square_values
plt.plot(square_values)

#show the plot
plt.show()
```

```python
# How to change the label type and graph thickness

square_values = [1, 4 , 9 , 16 , 25 , 36 , 49 , 64 , 81 ]

#plot the square_values
plt.plot(square_values, linewidth = 3)

#Setting the chart title and the label axis 
plt.title("Plot of Square Values", fontsize = 24)
plt.xlabel("Value", fontsize = 24)
plt.ylabel("Square of Value", fontsize = 14)

#Setting the tick label size 
plt.tick_params (axis = 'both', labelsize = 14)

#show the plot
plt.show()
```

```python
#Setting and plotting the input and output values for the plot 

#setting the input values 
input_values = [1 , 2 , 3 , 4 , 5 , 6 , 7, 8 , 9 ]
square_values = [1, 4 , 9 , 16 , 25 , 36 , 49 , 64 , 81  ]

plt.plot(input_values, square_values, linewidth = 4)

#Setting the chart title and the label axis 
plt.title("Plot of Square Values", fontsize = 24)
plt.xlabel("Value", fontsize = 24)
plt.ylabel("Square of Value", fontsize = 14)

#Setting the tick label size 
plt.tick_params (axis = 'both', labelsize = 14)

#show the plot
plt.show()
```

```python
# Plotting Individual Points with scatter ()

#use the scatter() function to plot individual points 
plt.scatter(2,4)
plt.show()
```

```python
plt.scatter ( 2 , 4 , s = 200 )

#Setting the chart title and the label axis
plt.title("Square Number Plot", fontsize = 20)
plt.xlabel("Value", fontsize = 16)
plt.ylabel("Square of values" , fontsize = 16)

#tick param label 
plt.tick_params(axis = 'both', which = 'major' , labelsize = 15)

#show the plot 
plt.show()
```

```python
# Plotting a series of individual points 

x_values = [2, 4, 6, 8]
y_values = [4, 16, 36, 64]

plt.scatter(x_values , y_values , s = 100 )
plt.show()
```

```python
# Calculating the data automatically

values_of_x = list( range  ( 1 , 1001) )
values_of_y = [ x**3 for x in values_of_x ]

plt.scatter(values_of_x, values_of_y, edgecolor = 'none' , c = 'orange', s = 30)
plt.show()
```
